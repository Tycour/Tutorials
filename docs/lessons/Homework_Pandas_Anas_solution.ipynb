```python
import numpy as np
import pandas as pd
```

```python
# read the file with expressions
df = pd.read_csv('_data\FPKM_pops_means.hisat2_fc_deseq2.tsv', sep='\t')
df
```

```python
# remove TXCHROM and TXSTRAND columns
df = df.drop(['TXCHROM', 'TXSTRAND'], axis=1)
df
```

```python
# reshuffle the columns order
# in df.loc --> [row, column]
df = df.loc[:,['pre', 'start', 'meiosis', 'post', 'gene']]
df
```

```python
# sent gene name as index and remove gene name column
df.index = df['gene']
df = df.drop('gene', axis=1)
df
```

```python
# function to calculate tau value
# for the details refer to: https://academic.oup.com/bib/article/18/2/205/2562739
def calculate_tau(x):
    max_e = max(x)
    x_hats = []
    
    for tissue_e in x:
        if max_e > 0:
            x_hat = tissue_e / max_e
            x_hats.append(1 - x_hat)
        else:
            x_hats.append(1)
            
    tau = sum(x_hats) / (len(x) - 1)
    
    return tau
```

```python
# create a copy of dataframe
df_annotated = df.copy()
```

```python
# apply the function
df_annotated['tau'] = df_annotated.apply(calculate_tau, axis=1)
df_annotated
```

```python
# define which cell population genes are specific for
# we set two conditions:
#  - tau > 0.7
#  - specificity is defined by having the highest expression    
df_annotated['specific'] = df_annotated.loc[df_annotated['tau'] > 0.7,['pre', 'start', 'meiosis', 'post']].idxmax(axis=1)
df_annotated
```

```python
#2 Find genes that have the peak expression in meiosis
df_annotated.loc[(df_annotated['meiosis'] > df_annotated['pre']) & (df_annotated['meiosis'] > df_annotated['start']) & (df_annotated['meiosis'] > df_annotated['post']), :]

```

```python
# reading the file using space delimiter (' ') rather than tab delimiter (sep='\t')
df1 = pd.read_csv('_data\CT_pops_diffexpr_PRE_MEIOSIS.hisat2_fc_deseq2.tab', delimiter=' ')
df1

#selecting genes with <2.5 log change in PRE with respect to MEIOSIS and p_values <0.05
df1_m = df1.loc[(df1['log2FoldChange'] < -2.5) & (df1['pvalue'] < 0.05) , :]
df1_m

#dropping columns that confuse me
df1_m = df1_m.drop(['baseMean', 'lfcSE', 'stat', 'padj'], axis = 1)

#renaming columns so I can tell after merging which one belongs to which dataset
df1_m = df1_m.rename(columns={"log2FoldChange": "log2FC_1", "pvalue": "pvalue_1"})
df1_m

```

```python
df2 = pd.read_csv('_data\CT_pops_diffexpr_START_MEIOSIS.hisat2_fc_deseq2.tab', delimiter=' ')
df2

#selecting genes with <2.5 log change in START with respect to MEIOSIS and p_values <0.05
df2_m = df2.loc[(df2['log2FoldChange'] < -2.5) & (df2['pvalue'] < 0.05) , :]
df2_m

#dropping columns that confuse me
df2_m = df2_m.drop(['baseMean', 'lfcSE', 'stat', 'padj'], axis = 1)

#renaming columns so I can tell after merging which one belongs to which dataset
df2_m = df2_m.rename(columns={"log2FoldChange": "log2FC_2", "pvalue": "pvalue_2"})
df2_m


```

```python
df3 = pd.read_csv('_data\CT_pops_diffexpr_MEIOSIS_POST.hisat2_fc_deseq2.tab', delimiter=' ')
df3

#selecting genes with >2.5 log change in MEIOSIS with respect to POST and p_values <0.05
df3_m = df3.loc[(df3['log2FoldChange'] > 2.5) & (df3['pvalue'] < 0.05) , :]
df3_m

#dropping columns that confuse me
df3_m = df3_m.drop(['baseMean', 'lfcSE', 'stat', 'padj'], axis = 1)

#renaming columns so I can tell after merging which one belongs to which dataset
df3_m = df3_m.rename(columns={"log2FoldChange": "log2FC_3", "pvalue": "pvalue_3"})
df3_m
```
