```python
### Visualise Overview Alignment

import dna_features_viewer

from dna_features_viewer import GraphicFeature, GraphicRecord
features=[
    GraphicFeature(start=0, end=27046, strand=0, color="#ffff33", fontdict={'size': '15'}, thickness=20,
                   label="Nanopore LSK*"),
    GraphicFeature(start=13316, end=13778, strand=0, color="#a85ab8", fontdict={'size': '15'}, thickness=18,
                   label="5' Inverse PCR"),
    GraphicFeature(start=13741, end=18913, strand=0, color="#ffcccc", fontdict={'size': '15'}, thickness=20,
                   label="Illumina"),
    GraphicFeature(start=13653, end=22095, strand=0, color="#f57242", fontdict={'size': '15'}, thickness=20,
                   label="Nanopore LSK* + Illumina"),
    GraphicFeature(start=17727, end=21820, strand=0, color="#ccccff", fontdict={'size': '15'}, thickness=20,
                   label="Nanopore LSK"),
    GraphicFeature(start=16896, end=21571, strand=0, color="#ccccff", fontdict={'size': '15'}, thickness=20,
                   label="Nanopore Cas9"),
    GraphicFeature(start=18906, end=19326, strand=0, color="#2aa134", fontdict={'size': '15'}, thickness=18,
                   label="3' Inverse PCR")
]
record = GraphicRecord(sequence_length=27046, features=features)
record.plot(figure_width=20)
```

```python
### Visualise 5' Crop

from dna_features_viewer import GraphicFeature, GraphicRecord
import matplotlib.pyplot as plt

features=[
    GraphicFeature(start=0, end=27046, strand=0, color="#ffd700",
                   label="tig00025588"),
    GraphicFeature(start=13316, end=13745, strand=-1, color="#a85ab8",
                   label="5' Galizi"),
    GraphicFeature(start=13741, end=18913, strand=+1, color="#ffcccc",
                   label="Ag(PMB)1"),
    GraphicFeature(start=13653, end=22095, strand=0, color="#f57242",
                   label="Node196"),
    GraphicFeature(start=13414, end=13778, strand=-1, color="#a85ab8",
                   label="5' New"),
    GraphicFeature(start=17727, end=21820, strand=+1, color="#ccccff",
                   label="Nanopore LSK"),
    GraphicFeature(start=16896, end=21571, strand=+1, color="#ccccff",
                   label="Nanopore Cas9"),
    GraphicFeature(start=18942, end=19326, strand=+1, color="#2aa134",
                   label="3' Galizi"),
    GraphicFeature(start=18906, end=19319, strand=+1, color="#2aa134",
                   label="3' New")]

record = GraphicRecord(sequence_length=27046, features=features)

zoom_start, zoom_end = 13200, 13900
cropped_record = record.crop((zoom_start, zoom_end))

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(30, 5))

# PLOT THE WHOLE SEQUENCE

ax1.set_title("Whole sequence", loc='left', weight='bold')
record.plot(ax=ax1)
ax1.fill_between((zoom_start, zoom_end), +1000, -1000, alpha=0.15)

# PLOT THE SEQUENCE DETAILS

cropped_record.plot(ax=ax2, plot_sequence=False)
ax2.set_title("5' detail", loc='left', weight='bold')

fig.savefig("5'_overview_and_detail.png", bbox_inches='tight')
```
